```python
import random

m = 50  # Size of matrices
n = 50


def generate_matrix():  # Generating matrices for tests
    global m
    global n
    min_limit = -250  # Lower limit of random values
    max_limit = 1000 + 15  # High limit of random values
    return [[random.randrange(min_limit, max_limit + 1) for y in range(m)] for x in range(n)]

matrix_second = generate_matrix()
for i in range(n):
    print(matrix_second[i])
```

### Задание №3

## Стандартная функция сортировки

```python
import time

matrix_standart = generate_matrix()
print('Autogenerated matrix for standart sort: ')
for i in range(n):
    print(matrix_standart[i])
time_standart = time.time()
for i in range(n):
    matrix_standart[i].sort()
standart_stop = time.time() - time_standart
print('\nSorted matrix by standart function: ')
for i in range(n):
    print(matrix_standart[i])
print("\nMatrix was sorted by standart Python sort in {0} s".format(standart_stop))
```

## Сортировка выбором

```python
def selection(str_array):  # Selection sort function
    for el in range(len(str_array) - 1):
        small = el
        for j in range(el + 1, len(str_array)):
            if str_array[j] < str_array[small]:
                small = j
        str_array[el], str_array[small] = str_array[small], str_array[el]
    return str_array


matrix_selection = generate_matrix()
print('Autogenerated matrix for selection sort: ')
for i in range(n):
    print(matrix_selection[i])
time_selection = time.monotonic()
for i in range(n):
    matrix_selection[i] = selection(matrix_selection[i])
time_stop = time.monotonic() - time_selection
print('\nSorted matrix by selection: ')
for i in range(n):
    print(matrix_selection[i])
print("\nMatrix was sorted by selection in {0} s".format(time_stop))
```

## Сортировка вставкой

```python
def insertion(str_array):  # Insertion sort function
    for i in range(1, len(str_array)):
        temp = str_array[i]
        j = i - 1
        while j >= 0 and temp < str_array[j]:
            str_array[j + 1] = str_array[j]
            j = j - 1
        str_array[j + 1] = temp
    return str_array


matrix_insertion = generate_matrix()
print('Autogenerated matrix for insertion sort: ')
for i in range(n):
    print(matrix_insertion[i])
time_insertion = time.monotonic()
for i in range(n):
    matrix_insertion[i] = insertion(matrix_insertion[i])
time_stop = time.monotonic() - time_insertion
print('\nSorted matrix by insertion: ')
for i in range(n):
    print(matrix_insertion[i])
print("\nMatrix was sorted by insertion in {0} s".format(time_stop))
```

## Сортировка обменом (пузырьком)

```python
def bubble(str_array):  # Bubble sort function
    for i in range(len(str_array) - 1):
        for j in range(len(str_array) - i - 1):
            if str_array[j] > str_array[j + 1]:
                str_array[j], str_array[j + 1] = str_array[j + 1], str_array[j]
    return str_array

matrix_bubble = generate_matrix()
print('Autogenerated matrix for bubble sort: ')
for i in range(n):
    print(matrix_bubble[i])
time_bubble = time.monotonic()
for i in range(n):
    matrix_bubble[i] = bubble(matrix_bubble[i])
time_stop = time.monotonic() - time_bubble
print('\nSorted matrix by bubble: ')
for i in range(n):
    print(matrix_bubble[i])
print("\nMatrix was sorted by bubble in {0} s".format(time_stop))
```

## Сортировка Шелла

```python
def shell(str_array):  # Shell sort function
    for i in range(len(str_array)):
        j = i - 1
        key = str_array[i]
        while str_array[j] > key and j >= 0:
            str_array[j + 1] = str_array[j]
            j -= 1
        str_array[j + 1] = key
    return str_array


matrix_shell = generate_matrix()
print('Autogenerated matrix for Shell sort: ')
for i in range(n):
    print(matrix_shell[i])
time_shell = time.monotonic()
for i in range(n):
    matrix_shell[i] = shell(matrix_shell[i])
time_stop = time.monotonic() - time_shell
print('\nSorted matrix by Shell: ')
for i in range(n):
    print(matrix_shell[i])
print("\nMatrix was sorted by Shell in {0} s".format(time_stop))
```

## Турнирная сортировка

```python
def heapify(heap_array, n, k):  # Tournament sort helper function
    m = k
    left = 2 * k
    right = 2 * k + 1
    if left < n and heap_array[m] < heap_array[left]:
        m = left
    if right < n and heap_array[m] < heap_array[right]:
        m = right
    if m != k:
        heap_array[k], heap_array[m] = heap_array[m], heap_array[k]
        heapify(heap_array, n, m)
    return heap_array

def tournament(str_array):  # tournament sort function
    for i in range(len(str_array) // 2, -1, -1):
        heapify(str_array, len(str_array), i)
    for i in range(len(str_array) - 1, -1, -1):
        if str_array[0] > str_array[i]:
            str_array[0], str_array[i] = str_array[i], str_array[0]
            heapify(str_array, i, 0)
    return str_array


matrix_tournament = generate_matrix()
print('Autogenerated matrix for tournament sort: ')
for i in range(n):
    print(matrix_tournament[i])
time_tournament = time.monotonic()
for i in range(n):
    matrix_tournament[i] = tournament(matrix_tournament[i])
time_stop = time.monotonic() - time_tournament
print('\nSorted matrix by tournament: ')
for i in range(n):
    print(matrix_tournament[i])
print("\nMatrix was sorted by tournament in {0} s".format(time_stop))
```

## Быстрая сортировка

```python
def quick(str_array):  # Quick sort function
    if len(str_array) <= 1:
        return str_array
    else:
        q = random.choice(str_array)
        lower = []
        middle = []
        higher = []
        for el in str_array:
            if el < q:
                lower.append(el)
            elif el > q:
                higher.append(el)
            else:
                middle.append(el)
        return quick(lower) + middle + quick(higher)

matrix_quick = generate_matrix()
print('Autogenerated matrix for quick sort: ')
for i in range(n):
    print(matrix_quick[i])
time_quick = time.monotonic()
for i in range(n):
    matrix_quick[i] = quick(matrix_quick[i])
time_stop = time.monotonic() - time_quick
print('\nSorted matrix by quick sort: ')
for i in range(n):
    print(matrix_quick[i])
print("\nMatrix was sorted by quick sort in {0} s".format(time_stop))
```

## Пирамидальная сортировка

```python
def heapify(heap_array, n, k):  # Heap sort helper function
    largest = k
    left = 2 * k + 1
    right = 2 * k + 2
    if left < n and heap_array[largest] < heap_array[left]:
        largest = left
    if right < n and heap_array[largest] < heap_array[right]:
        largest = right
    if largest != k:
        heap_array[k], heap_array[largest] = heap_array[largest], heap_array[k]
        heapify(heap_array, n, largest)
    return heap_array

def heap(str_array):  # Heap sort function
    for i in range(len(str_array), -1, -1):
        heapify(str_array, len(str_array), i)
    for i in range(len(str_array) - 1, 0, -1):
        str_array[i], str_array[0] = str_array[0], str_array[i]
        heapify(str_array, i, 0)
    return str_array

matrix_heap = generate_matrix()
print('Autogenerated matrix for heap sort: ')
for i in range(n):
    print(matrix_heap[i])
time_heap = time.monotonic()
for i in range(n):
    matrix_heap[i] = heap(matrix_heap[i])
time_stop = time.monotonic() - time_heap
print('\nSorted matrix by heap: ')
for i in range(n):
    print(matrix_heap[i])
print("\nMatrix was sorted by heap in {0} s".format(time_stop))
```
